```python
import pandas as pd
import numpy as np
```

```python
data_pop = pd.read_excel('PopulationEstimates.xls')
data_pop = data_pop.dropna(subset=['Rural-urban_Continuum Code_2013', 'Urban_Influence_Code_2013', 'FIPStxt'])
data_pop = data_pop.loc[:, ['FIPStxt', 'State', 'Area_Name']].drop_duplicates()
```

```python
my_dict = data_pop
len(my_dict)
```

```python
data_pov = pd.read_excel('PovertyEstimates.xls')
data_pov = data_pov.dropna(subset=['Rural-urban_Continuum_Code_2013', 'Urban_Influence_Code_2013', 'FIPStxt'])
data_pov = data_pov.loc[:, ['FIPStxt', 'Stabr', 'Area_name']].drop_duplicates()
data_pov = data_pov.rename(columns=dict(zip(data_pov.columns, my_dict.columns)))
#data_pov
mask = list(~data_pov['FIPStxt'].isin(my_dict['FIPStxt']))
my_dict = my_dict.append(data_pov.loc[mask,:], ignore_index=True)
```

```python
len(my_dict)
```

```python
data_ed = pd.read_excel('Education.xls')
data_ed = data_ed.dropna(subset=['2013 Rural-urban Continuum Code', '2013 Urban Influence Code', 'FIPS Code'])
data_ed = data_ed.loc[:, ['FIPS Code', 'State', 'Area name']].drop_duplicates()
data_ed = data_ed.rename(columns=dict(zip(data_ed.columns, my_dict.columns)))
data_ed
mask = list(~data_ed['FIPStxt'].isin(my_dict['FIPStxt']))
my_dict = my_dict.append(data_ed.loc[mask,:], ignore_index=True)
```

```python
data_em = pd.read_excel('Unemployment.xls')
data_em = data_em.dropna(subset=['Rural_urban_continuum_code_2013', 'Urban_influence_code_2013', 'fips_txt'])
data_em = data_em.loc[:,['fips_txt', 'Stabr', 'area_name']].drop_duplicates()
data_em = data_em.rename(columns=dict(zip(data_em.columns, my_dict.columns)))
mask = list(~data_em['FIPStxt'].isin(my_dict['FIPStxt']))
my_dict = my_dict.append(data_em.loc[mask,:], ignore_index=True)
#data_em
```
